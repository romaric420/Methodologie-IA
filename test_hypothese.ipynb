```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt 
from scipy.stats import binomtest,chisquare,ttest_1samp
import seaborn as sns 
```

```python
df = sns.load_dataset('tips')
df.head()
```

```python
p = 0.5

df['time'].value_counts(normalize=True)

k = df["time"].value_counts()['Dinner']

n=len(df)

alpha = 0.2

p_value = binomtest(k=k,n=n,p=p).pvalue

if p_value < alpha:
    print("nous avons suffisament devidences pour rejetter h0")
else:
    print("nous navons pas suffisament d'evidences pour rejeter H0")
```

```python

obs = df['day'].value_counts(normalize=True, sort=False).round(2).values
obs = obs * len(df) 

fre_exp = np.array([0.2, 0.1, 0.4, 0.3]) * len(df)

alpha  = 0.2

p_value = chisquare(f_obs=obs, f_exp=fre_exp).pvalue
print(f"Valeur p : {p_value}")

if p_value < alpha:
    print('nous avons suffisament delements pour rejetes h0')
else:
    print("nous navons pas suffisament dhipothese pour rejetter h0")
```
