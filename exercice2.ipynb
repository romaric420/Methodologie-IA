```python
sns.scatterplot(data=df, x='mpg', y='horsepower', hue='model_year')

```

```python
temp_df = df.groupby("model_year")[["mpg","horsepower"]].mean()

plt.scatter(temp_df["mpg"], temp_df["horsepower"])


for year in temp_df.index:
    plt.text(temp_df.loc[year, "mpg"], 
             temp_df.loc[year, "horsepower"], 
             str(int(year)),  
             fontsize=9, 
             ha='center',  
             va='bottom')   

plt.xlabel("MPG")
plt.ylabel("Horsepower")
plt.show()
```
