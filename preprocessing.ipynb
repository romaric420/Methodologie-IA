# OnehotEncoder()

```python
df = sns.load_dataset("diamonds")
df.head()
```

```python
df['clarity'].unique()
```

```python
cut_oder =['very bad','bad','Fair','Good','Very Good','Premium','Ideal']
color_order =['J','I','H','G','F','E','D']
clarity_order =['I1','SI2','SI1','VS2','VS1','VVS2','VVS1','IF']

encoder = OrdinalEncoder(
    categories=[cut_oder,color_order,clarity_order],
    handle_unknown="use_encoded_value",
    unknown_value=-1,                     
                         )
encoder.fit_transform(df[["cut","color","clarity"]])
```

```python
encoder.transform([['bad','G','SI1']])
```
